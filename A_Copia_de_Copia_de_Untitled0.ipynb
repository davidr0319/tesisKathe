```python
# Importando conjunto de datos
import pandas as pd
import numpy as np

#Cargando datos pima indians dataset
data = pd.read_excel(r"/content/drive/MyDrive/final/TRABAJO FINAL/A cancer patient data sets.xlsx")
data.head()
```

```python
# Quito campos indeseados
data=data.drop(columns=["index","Patient Id","Age"],axis=1)
#resumen de los datos
data.head()
```

```python
data.describe()
```

```python
# Ajusto valores de etiquetas
Salida_Strategy={"Low":1,"High":2}
data["Level"]=data["Level"].map(Salida_Strategy)

#resumen de los datos
data.head()
```

```python
# Dividir el conjunto de datos en variables de entrada (X) y salida (Y)
Datos=np.array(data)
X = Datos[:,0:22] #Entrada
Y = Datos[:,22] #Salida
print(X.shape,Y.shape)
```

```python
#3. Divida la entrada y la salida en conjuntos de entrenamiento y prueba
import sklearn
from sklearn.model_selection import train_test_split
X_train, X_test,Y_train, Y_test= train_test_split(X,Y,test_size=0.1)
print(X_train.shape,X_test.shape )
```

=========================Vecinos m√°s cercanos==============================

```python
#4. Evaluando casos mediante VECINOS
from sklearn.neighbors import KNeighborsClassifier
pronostico_0 = KNeighborsClassifier(5) #
pronostico_0.fit(X_train,Y_train)
Y_pred_0 =pronostico_0.predict (X_test)
# calculando desempe√±o (Exactitud)
from sklearn.metrics import accuracy_score,precision_score
print("Accuracy ",accuracy_score(Y_test, Y_pred_0))
```

```python
#Guardando el pronostico entrenado
import joblib
joblib.dump(pronostico_0, "/content/drive/MyDrive/final/TRABAJO FINAL/pronostico_VECINOS.joblib")
```

=========================Bayes==============================

```python
#5. Evaluando casos mediante Naive Bayes
from sklearn.naive_bayes import GaussianNB
pronostico_1 = GaussianNB()
pronostico_1.fit(X_train, Y_train)
Y_pred =pronostico_1.predict (X_test)
from sklearn.metrics import accuracy_score,precision_score
print("Accuracy ",accuracy_score(Y_test, Y_pred))
```

```python
#Guardando el pronostico entrenado
import joblib
joblib.dump(pronostico_1, "/content/drive/MyDrive/final/TRABAJO FINAL/pronostico_Bayes.joblib") 
```

======================LDA=================================

```python
#6. Evaluando casos mediante LDA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
pronostico_2 = LinearDiscriminantAnalysis()
pronostico_2.fit(X_train, Y_train)
Y_pred_2 =pronostico_2.predict (X_test)
from sklearn.metrics import accuracy_score,precision_score
print("Accuracy ",accuracy_score(Y_test, Y_pred_2))
```

```python
#Guardando el pronostico entrenado LDA
import joblib
joblib.dump(pronostico_2, "/content/drive/MyDrive/final/TRABAJO FINAL/pronostico_LDA.joblib") 
```

==========================QDA==================================

```python
#8. Evaluando casos mediante QDA
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
pronostico_3 = QuadraticDiscriminantAnalysis()
pronostico_3.fit(X_train, Y_train)
Y_pred_3 =pronostico_3.predict (X_test)
from sklearn.metrics import accuracy_score,precision_score
print("Accuracy ",accuracy_score(Y_test, Y_pred_3))
```

```python
#Guardando el pronostico entrenado
import joblib
joblib.dump(pronostico_3, "/content/drive/MyDrive/final/TRABAJO FINAL/pronostico_QDA.joblib") 
```

=============================Tree===============================

```python
#9. Evaluando casos mediante Tree 
from sklearn.tree import DecisionTreeClassifier
pronostico_4 = DecisionTreeClassifier()
pronostico_4.fit(X_train, Y_train)
Y_pred_4 =pronostico_4.predict (X_test)
from sklearn.metrics import accuracy_score,precision_score
print("Accuracy ",accuracy_score(Y_test, Y_pred_4))
```

```python
#Guardando el pronostico entrenado
import joblib
joblib.dump(pronostico_4, "/content/drive/MyDrive/final/TRABAJO FINAL/pronostico_TREE.joblib") 
```
